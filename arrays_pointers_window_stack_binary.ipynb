```python
def is_anagram_brute(t,s):

    #Anagram needs to have the same number of letters
    if len(t)!=len(s):
        return False

    elements = {}

    for i in range(len(t)):
        if t[i] in elements:
            elements[t[i]]= elements[t[i]]+1
        else: 
            elements[t[i]]=1

    for j in range(len(s)):
        if s[j] in elements:
            #Remove key if 
            elements[s[j]]= elements[t[i]]-1

        else: 
            return False

    if sum(elements.values())==0: 
        return True 

    else: 
        return False
```

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

https://leetcode.com/problems/container-with-most-water/

```python
class Solution:

  def maxArea_brute(self, height: List[int]) -> int:

    max_area = 0
    for i in range(len(height)):
      for j in range(len(height)):
        if(j>i):
          temp_area = (j-i)* min(height[i],height[j])
          if temp_area > max_area:
            max_area = temp_area

    return max_area

  def maxArea(self, height):

    i, j = 0, len(height)-1 
    
    max_area = 0

    while i<j: 
      if (min(height[i], height[j]) * (i-j)) > max_area: 
        max_area = (min(height[i], height[j]) * (i-j))


      if min(height[i], height[j]) == height[j]: 
        j-=1 

      else: 
        i+=1


    return max_area
```

## 20. Valid parentheses 

Given a string s containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.

An input string is valid if:

- Open brackets must be closed by the same type of brackets.
- Open brackets must be closed in the correct order.
- Every close bracket has a corresponding open bracket of the same type.


```python
def valid_parenthesis(s):

    stack = []
    open_set = ('(','{', '[')

    for element in s: 

        if element in open_set:
            stack.append(element)

        elif element==')':
            if stack[-1]!='(':
                return False
                
            else: 
                stack.pop()

        elif element==']':
            if stack[-1]!='[':
                return False 
                
            else: 
                stack.pop()

        elif element=='}':
            if stack[-1]!='{':
                return False 
                
            else: 
                stack.pop()

    if len(stack)==0:
        return True 

    else: 
        return False

```

```python
## Slightly more optimised code
class MinStack:
    def __init__(self):
        self.stack = []
        self.minStack = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        val = min(val, self.minStack[-1] if self.minStack else val)
        self.minStack.append(val)

    def pop(self) -> None:
        self.stack.pop()
        self.minStack.pop()

    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.minStack[-1]
```

## 704. Binary search 

Given an array of integers nums which is sorted in ascending order, and an integer target, write a function to search target in nums. If target exists, then return its index. Otherwise, return -1.

You must write an algorithm with O(log n) runtime complexity.

```python
def binary_search(nums, target): 
    
    l,m, r = 0, int(len(nums)/2), len(nums)-1

    while l<=r:

        if nums[m]==target: 
            return m 

        elif nums[m]<target: 
            l = m
            m = m+ int((r-l)/2)
        
        else:
            r = m 
            m = m - int((r-l)/2)

    return -1

```

```python
nums = [-1,0,3,5,9,12]

target = 9

binary_search(nums,target)
```
